### Installing and loading packages

```r
# install.packages('rgexf')
# install.packages('igraph')
# install.packages('visNetwork')
library(rgexf)
library(igraph)
library(visNetwork)
```

### Graph creating and visualization

```r
graph = read.gexf("att_more_than_5.txt") # file reading

data_igraph <- gexf.to.igraph(graph) # graph creating

data_igraph # graph description
plot(data_igraph, vertex.label=NA, vertex_size=0.1, layout = layout_with_lgl)
visIgraph(data_igraph, layout = "layout_with_fr") %>%
visOptions(highlightNearest = list(enabled = T, hover = T), 
            nodesIdSelection = T) # visualisation
```

### Graph characteristics

```r
igraph::components(data_igraph) # components of connectivity (a maximal connected subgraph of the graph)
num_edges <- E(data_igraph)
edge_density(data_igraph) # graph density
```

#### Diameter

```r
diameter(data_igraph) # diameter
diam <- get_diameter(data_igraph)
diam # paths corresponding to the diameter
```

#### Distances between nodes

```r
distances(data_igraph) # a matrix of the shortest paths
mean_distance(data_igraph) # the average of the shortest paths
```

#### Nodes degrees

```r
deg <- degree(data_igraph) 
deg # the centrality of the vertex of the graph by the degree of vertex
```

```r
average_degree <- mean(degree(data_igraph)) # average degree of the vertexes
average_degree
is_connected <- is_connected(data_igraph) # the connectivity of the graph
is_connected
```

```r
par(mar = c(1, 4, 2, 0))
hist(deg, main = "Histogram of node degree", breaks = 1:vcount(data_igraph)-1, prob = TRUE) # histogram of vertex distribution
xfit <- seq(min(deg), max(deg), length = 40) 
yfit <- dnorm(xfit, mean = mean(deg), sd = sd(deg))
lines(xfit, yfit, col = "black", lwd = 2) 
```

#### Centrality

```r
clos <- closeness(data_igraph) # the centrality of the vertex in proximity
betw <- betweenness(data_igraph) # betwenness centrality
centr <- centr_betw(data_igraph) # shortest path centrality
eig <- eigen_centrality(data_igraph) # vertex eigenvalues
eig_centrality <- centr_eigen(data_igraph) # eigenvalue centrality

# combine all centrality into one matrix and see the correlations:
df <- data.frame (deg, clos, betw, eig$vector)
cor(df) # correlation matrix
```

### Erdős-Rényi model

```r
p <- edge_density(data_igraph)
g.random <- erdos.renyi.game(n = gorder(data_igraph), p.or.m = p, type = "gnp")
par(mfrow = c(1, 2))
plot(data_igraph, vertex.label=NA, vertex_size=15, layout = layout_nicely) # graph visualisation
plot(g.random, vertex.label=NA, layout = layout_nicely, vertex.size = 5) # random Erdős-Rényi model graph visualisation
```

```r
par(mfrow = c(1, 2))
hist(degree(data_igraph), main = "Histogram of node degree", breaks = 1:vcount(data_igraph)-1, prob = TRUE)
hist(degree(g.random), main = "Histogram of node degree", breaks = 1:vcount(data_igraph)-1, prob = TRUE)
par(mfrow = c(1, 1))
```

```r
gl <- vector('list', 1000)
m <- rep(NA, 1000)
for(i in 1:1000){
  gl[[i]] <- erdos.renyi.game(n = gorder(data_igraph), p.or.m = p, type = "gnp")
  m[i] <- mean_distance(gl[[i]])
}
hist(m, xlim = range(c(min(m), 6)))
abline(v = mean_distance(data_igraph), col = "red", lty = 3, lwd = 2)
```

### Watts-Strogatz model

### Barabási-Albert model

```r
g.BA <- sample_pa(n = gorder(data_igraph), power = 1, directed = FALSE)
hist(degree(g.BA), main = "Histogram of node degree", breaks = 1:vcount(data_igraph)-1, prob = TRUE)
```

```r
g <- sample_pa(500, power = 1, directed = FALSE)
plot(g, vertex.label=NA, vertex.size = 2)
plot(degree_distribution(g)) 
plot(degree_distribution(g), log='xy')
```

```r
g <- sample_pa(500, out.dist = c(0.25, 0.5, 0.25), directed = FALSE, zero.appeal = 1)
plot(g, vertex.label=NA, vertex.size = degree(g)/2)
plot(degree_distribution(g))
plot(degree_distribution(g), log='xy')
```

### Assortativity

```r
p <- edge_density(data_igraph)

assortativity_degree(data_igraph) # general assortativity
vert_attr <- vertex_attr(data_igraph) # vertex graph attributes
vertex_attr_names(data_igraph) # vertex graph attributes names
assortativity_nominal(data_igraph, types = V(data_igraph)$name, directed = FALSE) # assortativity by name
assortativity(data_igraph, values = V(data_igraph)$name, directed = FALSE) # assortativity by name

gl <- vector('list', 1000)
m <- rep(NA, 1000)
for(i in 1:1000){
  gl[[i]] <- erdos.renyi.game(n = vcount(data_igraph), p.or.m = p, type = "gnp")
  m[i] <- assortativity(gl[[i]], values = sample(V(data_igraph)$name))
}
hist(m, xlim = range(c(-1, 1)))
abline(v = assortativity(data_igraph, values = V(data_igraph)$name), col = "pink", lty = 3, lwd = 2)
```

### Dyadicity and heterophilicity

```r
V(data_igraph)$color <- seq(from = 1, to = 236)
edgeList <- as_edgelist(data_igraph)
faction <- matrix(c(V(data_igraph)$color, V(data_igraph)$name), nrow = 236, ncol = 2)
FromLabel <- as.numeric(faction[match(as.numeric(edgeList[,1]), as.numeric(faction[,1])), 2])
ToLabel <- as.numeric(faction[match(as.numeric(edgeList[,2]), as.numeric(faction[,1])), 2]) 
edgeType <- FromLabel + ToLabel

table(edgeType)
table(as.numeric(faction[,2]))
exp_dyad <- table(as.numeric(faction[,2]))[1] * (table(as.numeric(faction[,2]))[1] - 1) * p/2
dyadicity <- table(edgeType)[1] / exp_dyad
exp_hetero <- table(as.numeric(faction[,2]))[1] * table(as.numeric(faction[,2]))[2] * p
heterophilicity <- table(edgeType)[2] / exp_hetero
```

### Transitivity

```r
local_transitivity <- transitivity(data_igraph, type = "local") # the line writes out how close a particular vertex is to forming cliques with its neighbors (measured from 0 to 1, 0 when there are no connections with neighbors, 1 when all are) 
global_transitivity <- transitivity(data_igraph, type = "global") # global cluster coefficient shows what fraction of the triples are closed
global_transitivity # We have 0.47, which is quite a lot

gl <- vector('list', 1000)
m <- rep(NA, 1000)
for(i in 1:1000){
  gl[[i]] <- erdos.renyi.game(n = vcount(data_igraph), p.or.m = p, type = "gnp")
  m[i] <- transitivity(gl[[i]], type = "global")
}
hist(m, xlim = range(c(0, 1)))
abline(v = transitivity(data_igraph, type = "global"), col = "blue2", lty = 3, lwd = 2)
```
